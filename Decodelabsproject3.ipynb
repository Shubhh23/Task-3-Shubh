```python
%%sql
SELECT COUNT(*) AS TotalOrders
FROM Orders;
```

Observation:
The Orders table contains 1200 records.

Insight:
The dataset consists of 1200 customer orders available for analysis.

```python
%%sql
SELECT AVG(TotalPrice) AS Averageordervalue FROM Orders;
```

Insight:
Customers spend approximately ₹1054 per order on average.

```python
%%sql 
SELECT SUM(TotalPrice) AS TotalRevenue FROM Orders;
```

Observation:
The total revenue generated from all orders is ₹12,64,761.96.
    
Insight:
The business generated approximately ₹12.65 lakh in revenue from 1200 customer orders.

```python
%%sql 
SELECT Product, COUNT(*) AS TotalOrders FROM Orders
GROUP BY Product;
```

```python
%%sql
SELECT Product,
COUNT(*) AS TotalOrders
FROM Orders
GROUP BY Product
ORDER BY TotalOrders DESC;
```

Observation:
Printer has the highest number of orders (181), while Phone has the lowest number of orders (156).

Insight:
Printer is the most popular product among customers, while Phone has the lowest demand in the dataset.

```python
%%sql 
SELECT*FROM Orders 
Where OrderStatus='Delivered';
```

Observation:
The query displays only orders with a Delivered status.
    
Insight:
Delivered orders represent successfully completed customer purchases and can be used to analyze fulfilled sales.

```python
%%sql
SELECT PaymentMethod, COUNT(*) AS TotalOrders FROM Orders
GROUP BY PaymentMethod
ORDER BY TotalOrders DESC;
```

Observation:
Online payment has the highest number of orders.
    
Insight:
Customers prefer online payment methods over other payment options.

Final Insights

1. The dataset contains 1200 customer orders.

2. The average order value is ₹1053.97.

3. Total revenue generated is ₹12,64,761.96.

4. Printer is the most ordered product, while Phone is the least ordered.

5. Product demand varies across categories, with Printer and Tablet showing the highest popularity.

6. Delivered orders represent successfully completed transactions.

7. SQL queries such as COUNT, AVG, SUM, GROUP BY, ORDER BY, and WHERE helped uncover key business insights from the dataset.
